# **Data Preprocessing**

**Missing Value**

Jakarta

```python
# Merapihkan data dengan memotong baris ke 121 ke bawah karena berisi NaN
dfJ = dfJ.loc[0:120]
dfJ = dfJ[['created_at', 'LT', 'LB', 'KT', 'KM', 'garasi', 'carport', 'lokasi', 'sertifikat', 'listrik', 'hadap', 'harga', 'URL', 'deskripsi']]
dfJ
```

```python
# Membersihkan data menggunakan fillna
dfJ['garasi'].fillna(0, inplace=True)
dfJ['carport'].fillna(0, inplace=True)
```

```python
# Mengecek data yang kosong dengan isnull
dfJ.isnull().sum()
```

```python
# Mengecek data yang tidak kosong
dfJ.notnull().sum()
```

```python
# Mengecek apakah sudah berhasil atau belum
dfJ
```

sisanya yaitu variabel listrik dan hadap tidak dilakukan apa-apa terhadap missing value yang ada, karena jika kita mengganti NaN = 0, kemungkinan penjual rumah sengaja atau lupa menginput data. Maka dari itu kita tidak bisa sembarangan mengganti NaN dengan 0, dan tidak mungkin di zaman ini sebuah perusahaan menjual rumah tanpa ada listrik.

```python
dfJ.duplicated().sum()
```

```python
dfJ.mean().astype('int64')
```
